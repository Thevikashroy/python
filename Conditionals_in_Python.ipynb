```python
iit_shortlist_percentage = 90.0
candidate_percentage = float(input("Enter your percetage: "))

if candidate_percentage >= iit_shortlist_percentage:
  print("congratulations! you are selected to an IIT college")
elif candidate_percentage < iit_shortlist_percentage and candidate_percentage > 88:
  print("you are in the waiting list4")
else:
  print("sorry, you didnot met the selection criteria")
```

**Movie Ticket Pricing**

```python
age = int(input("Enter your age:"))
day = input("Is it a weekday or weekend ").lower()

if age < 5:
  price = 0
elif age <= 12:
  price = 100
elif age >= 65:
  price = 80
else:
  price = 200

## apply surplus charges for the weekend
if day == "weekend":
  price += 200

print("Movie ticket price",price ,"INR")
```

Palindrome true or false

```python
S = input("Enter the palindrome String: ")

n=S[::-1]
if S == n:
  print(True)
else:
  print(False)
```
